```python
#   ,     
#      .
#  ,     

d = {'A': ['a1', 'a2', 'a3'],
     'B': ['b1', 'b2', 'b3', 'b4']}

d_inverted = {}
for k in d.keys():
  for v in d[k]:
      d_inverted.update({v:k})

d_inverted
```

```python

cultures_grouped_inverted = {}
for k in cultures_grouped.keys():
  for v in cultures_grouped[k]:
    cultures_grouped_inverted.update({v:k})

cultures_grouped_inverted
```

```python
one_hot2 = pd.get_dummies(data['isAliveSpouse']).astype(int)
one_hot3 = pd.get_dummies(data['culture_grouped']).astype(int)
data=pd.concat([data,one_hot2],axis=1)
data=pd.concat([data,one_hot3],axis=1)
data.drop(columns=['title','culture_grouped'],inplace=True)
data['isAliveSpouse']=data['isAliveSpouse'].replace('isAliveSpouse_no_data',-1)
data['isAliveHeir']=data['isAliveHeir'].fillna('Heir_no_data')
one_hot66 = pd.get_dummies(data['isAliveHeir']).astype(int)
data=pd.concat([data,one_hot66],axis=1)
data['isAliveHeir']=data['isAliveHeir'].replace('Heir_no_data',-1)
cors=data.corr()
f,ax=plt.subplots(figsize=(12,9))
cols = cors.nlargest(20, 'isAlive')['isAlive'].index
cm = np.corrcoef(data[cols].values.T)
sns.set(font_scale=1.25)
sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 20}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
